#  Problem: Binary Classification for Email Spam Detection

Given a dataset of emails with labeled spam or non-spam, use a perceptron to classify whether a new email is spam or not.

```python
# Import Libraries
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('emails.csv')
dataset.head()
```

```python
# Statistical Analysis and data cleaning

df = dataset.copy()
df.shape
```

```python
df.isnull().count()
```

```python
# Dropping null values
df = df.dropna()
```

```python
df.info()
df.iloc[: , 2:3001]
```

```python
# Splliting into training and testing

from sklearn.model_selection import train_test_split

features = df.iloc[ : , 1 : 3001]
target = df[['Prediction']]
```

```python
features.head()
```

```python
target.head()
```

```python
X = features
X.head()
```

```python
Y = target
Y.head()
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, train_size = 0.8, test_size = 0.2, random_state = 42)
```

```python
# Apply perceptron

from sklearn.linear_model import Perceptron

perceptron = Perceptron()
perceptron.fit(X_train,Y_train)
y_pred = perceptron.predict(X_test)
y_pred = pd.DataFrame(y_pred)
```

```python
# finding accuracy:
from sklearn import metrics
mse = metrics.mean_squared_error(Y_test,y_pred)
print(f'Mean Squared Error: {mse}')
accuracy_score = metrics.accuracy_score(Y_test,y_pred)
print(f'Accuracy: {accuracy_score}')
```
